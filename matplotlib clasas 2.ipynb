```python
months = np.array(["jan", "frb", "march", "april", "may"])
profit = np.array([12, 23, 20, 23, 22])

col = ["red", "blue", "pink", "yellow", "green"]

wth = [0.3, 0.8, 0.6, 0.7, 0.5]

plt.bar(months, profit, color = col, width = wth, label = "My Revnute")

plt.grid()
plt.xlabel("Monthhs")
plt.ylabel("profit")
plt.title("Revenue")
plt.legend()
```

```python
months = np.array(["jan", "frb", "march", "april", "may"])
profit = np.array([12, 23, 20, 23, 22])

col = ["red", "blue", "pink", "yellow", "green"]

wth = [0.3, 0.8, 0.6, 0.7, 0.5]

plt.barh(months, profit, color = col, label = "My Revnute")
# plt.bar(months, profit, color = col, width = wth, label = "My Revnute")

# plt.grid()
plt.xlabel("Monthhs")
plt.ylabel("profit")
plt.title("Revenue")
plt.legend()
```

```python
months = np.array(["jan", "frb", "march", "april", "may"])
profit = np.array([12, 23, 20, 23, 22])

col = ["red", "blue", "pink", "yellow", "green"]

wth = [0.3, 0.8, 0.6, 0.7, 0.5]

plt.bar(months, profit, color = col, label = "My Revnute")
plt.plot(months, profit, label = "Trend Line", marker = ".", ms = 15, color = "black")

# plt.grid()
plt.xlabel("Monthhs")
plt.ylabel("profit")
plt.title("Revenue")
plt.legend()
```
