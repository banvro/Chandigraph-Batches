```python
monthname = np.array(["jan", "feb", "march", "april", "may", "June"])
sales = np.array([96, 80, 76, 90, 78, 90])
profit = np.array([70, 79, 89, 80, 67, 79])
```

```python
plt.plot(monthname, sales, marker = ".", ms = 16, mec = "red", 
         mfc = "yellow", 
         color = "green", 
        linewidth = 2, linestyle = "solid", label = "My Sales Trend Line")
plt.plot(monthname, profit, marker = "*", ms = 15, mec = "blue", mfc = "orange", label = "My Profit Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("My Sales Report", color = "red", loc = "left")
plt.grid(color = "green", linewidth = 1, linestyle = "dotted")
plt.legend()
```
